```python
import pandas as pd
df = pd.read_csv('Survey.csv')
```

```python
df
```

```python
# Dropping unncessary columns
df = df.drop(['Timestamp', 'Email Address', 'Your Name', 'Score', 'Unnamed: 11', 'Unnamed: 12', 'Registration Number'], axis=1)
```

```python
df
```

```python
# Droped this column because 99% rows consists of NAN values 
df = df.drop('Approximately, for how many hours your laptop lasts when not plugged in to power supply ?', axis = 1)
```

```python
df
```

```python
df = df.rename(columns = {'What is your Laptop Brand':'laptop_brand', 
               "For how many year's you are using this laptop ?" : "number_of_years_used", 
                "State the average battery life of your current Laptop in one complete charge" : "battery_life(in hrs)",
                "Number of hours you spend using your laptop" : "usage_per_day(in hrs)", 
                "Number of time you charge your laptop in a day ?": "number_of_times_charged_per_day"})
```

```python
df.head(3)
```

```python
df['battery_life(in hrs)'] = df['battery_life(in hrs)'].map({'less than 2': 0, 'less than 4': 1, 'less than 6': 2, 
                                                             'less than 8': 3})
df['usage_per_day(in hrs)'] = df['usage_per_day(in hrs)'].map({'less than 20 hours': 2, 'less than 10 hours': 1, 
                                                               'less than 5 hours': 0})
```

```python
df.head(3)
```

```python
df['laptop_brand'].value_counts()
```

```python
df['laptop_brand'] = df['laptop_brand'].map({'HP': 1, 'Lenovo': 2, 'Dell': 0, 'Others': 4, 'Asus':3})
```

```python
df['laptop_brand'].corr(df['battery_life(in hrs)'])
```

```python
df['usage_per_day(in hrs)'].corr(df['battery_life(in hrs)'])
```

```python
df['number_of_times_charged_per_day'].corr(df['battery_life(in hrs)'])
```

```python
df['number_of_times_charged_per_day'].corr(df['laptop_brand'])
```

```python
df['number_of_years_used'].corr(df['battery_life(in hrs)'])
```
