```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
fruits=pd.read_table(r'E:\Study\course3_downloads\fruit_data_with_colors.txt')
```

```python
X= fruits[['mass','width','height','color_score']]
y=fruits['fruit_label']
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
x_train
```

```python
x_train.shape
```

```python
from matplotlib import cm
cmap=cm.get_cmap('gnuplot')
scatter = pd.plotting.scatter_matrix(x_train,c=y_train, marker='o',s=40,hist_kwds={'bins':15},figsize=(12,12),cmap=cmap) 
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x_train['width'],x_train['height'],x_train['color_score'],c=y_train,marker='o',s=100)
ax.set_xlabel('width')
ax.set_ylabel('height')
ax.set_zlabel('color_score')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
fruit_prediction=knn.predict([[20,4.3, 5.5,.55]])
lookup_fruit_name[fruit_prediction[0]]
```
